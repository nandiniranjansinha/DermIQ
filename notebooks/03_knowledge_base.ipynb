```python
irritants = [
    'fragrance', 'parfum', 'alcohol denat', 'denatured alcohol',
    'limonene', 'linalool', 'citral', 'eugenol', 'geraniol',
    'cinnamal', 'benzyl alcohol', 'isopropyl alcohol',
    'menthol', 'sodium lauryl sulfate', 'citronellol'
]

comedogenic = [
    'coconut oil', 'cocoa butter', 'isopropyl myristate',
    'isopropyl palmitate', 'acetylated lanolin', 'wheat germ oil',
    'flaxseed oil', 'soybean oil', 'cottonseed oil'
]

harmful = [
    'parabens', 'methylparaben', 'propylparaben', 'butylparaben',
    'formaldehyde', 'dmdm hydantoin', 'quaternium-15',
    'triclosan', 'sodium lauryl sulfate', 'polyethylene',
    'butylated hydroxyanisole', 'bha', 'bht'
]

beneficial = [
    'niacinamide', 'retinol', 'retinyl palmitate', 'hyaluronic acid',
    'sodium hyaluronate', 'glycerin', 'ceramide np', 'ceramide ap',
    'ceramide eop', 'squalane', 'tocopherol', 'tocopheryl acetate',
    'ascorbic acid', 'vitamin c', 'panthenol', 'allantoin',
    'centella asiatica', 'salicylic acid', 'lactic acid',
    'glycolic acid', 'azelaic acid', 'caffeine', 'peptides',
    'collagen', 'aloe barbadensis leaf juice'
]
```

```python
import pandas as pd
import ast
```

```python
skincare_df = pd.read_csv('data/processed/skincare_df.csv')
```

```python
skincare_df['ingredient_list']
```

```python
def analyze_ingredients(ingredient_list):
    found_irritants = [i for i in irritants if i in ingredient_list]
    found_comedogenic = [i for i in comedogenic if i in ingredient_list]
    found_harmful = [i for i in harmful if i in ingredient_list]
    found_beneficial = [i for i in beneficial if i in ingredient_list]
    return {
        'irritants' : found_irritants,
        'comedogenic' : found_comedogenic,
        'harmful' : found_harmful,
        'beneficial' : found_beneficial
    }

result = analyze_ingredients(skincare_df['ingredient_list'].iloc[0])
print(result)
```

```python
skincare_df.head(2)
```

```python
# Apply analyze_ingredients() to dataset
analysis = skincare_df['ingredient_list'].apply(analyze_ingredients)

# Split it into separate columns
skincare_df['irritants_found'] = analysis.apply(lambda x:x['irritants'])
skincare_df['comedogenic_found'] = analysis.apply(lambda x:x['comedogenic'])
skincare_df['harmful_found'] = analysis.apply(lambda x:x['harmful'])
skincare_df['beneficial_found'] = analysis.apply(lambda x:x['beneficial'])

# Check result
skincare_df[['product_name','irritants_found','comedogenic_found','harmful_found','beneficial_found']].head(5)
```

```python
skincare_df.to_csv('data/processed/skincare_knowledge.csv', index = False)
print('Saved!')
print(skincare_df.shape)
```
