```python
import pandas as pd
import ast
```

```python
skincare_df = pd.read_csv("data/processed/skincare_knowledge.csv")
```

```python
skincare_df.columns
```

```python
skincare_df['irritants_found']
```

```python
# Convert list columns back to real list
skincare_df['ingredient_list'] = skincare_df['ingredient_list'].apply(ast.literal_eval)
skincare_df['irritants_found'] = skincare_df['irritants_found'].apply(ast.literal_eval)
skincare_df['comedogenic_found'] = skincare_df['comedogenic_found'].apply(ast.literal_eval)
skincare_df['harmful_found'] = skincare_df['harmful_found'].apply(ast.literal_eval)
skincare_df['beneficial_found'] = skincare_df['beneficial_found'].apply(ast.literal_eval)
```

```python
skincare_df['irritants_found']
```

```python
skincare_df.columns
```

```python
# Count features - how many of each type per product
skincare_df['total_ingredients'] = skincare_df['ingredient_list'].apply(len)
skincare_df['irritants_count'] = skincare_df['irritants_found'].apply(len)
skincare_df['comedogenic_count'] = skincare_df['comedogenic_found'].apply(len)
skincare_df['harmful_count'] = skincare_df['harmful_found'].apply(len)
skincare_df['beneficial_count'] = skincare_df['beneficial_found'].apply(len)
```

```python
skincare_df['total_ingredients']
```

```python
# Ratio features
skincare_df['irritants_ratio'] = skincare_df['irritants_count'] / skincare_df['total_ingredients']
skincare_df['comedogenic_ratio'] = skincare_df['comedogenic_count'] / skincare_df['total_ingredients']
skincare_df['harmful_ratio'] = skincare_df['harmful_count'] / skincare_df['total_ingredients']
skincare_df['beneficial_ratio'] = skincare_df['beneficial_count'] / skincare_df['total_ingredients']
```

```python
# Check results
skincare_df[['product_name','total_ingredients','irritants_count','irritants_ratio','beneficial_count','beneficial_ratio']].head()
```

```python
# Binary features - 0 or 1
skincare_df['has_fragrance'] = skincare_df['ingredient_list'].apply(lambda x:1 if 'fragrance' in x or 'parfum' in x else 0)
skincare_df['has_alcohol'] = skincare_df['ingredient_list'].apply(lambda x:1 if 'alcohol denat' in x else 0)

skincare_df['has_harmful'] = skincare_df['harmful_found'].apply(lambda x:1 if len(x) > 0 else 0)
skincare_df['has_beneficial'] = skincare_df['beneficial_found'].apply(lambda x: 1 if len(x) > 0 else 0)

skincare_df[['product_name', 'has_fragrance', 'has_alcohol', 
             'has_harmful', 'has_beneficial']].head()
```

```python
skincare_df
```

```python
skincare_df['highlights_list'] = skincare_df['highlights_list'].apply(ast.literal_eval)

# Target labels
skincare_df['is_for_dry_skin']  = skincare_df['highlights_list'].apply(lambda x: 1 if 'Best for Dry, Combo, Normal Skin' in x else 0)

skincare_df['is_for_oily_skin'] = skincare_df['highlights_list'].apply(lambda x: 1 if 'Best for Oily, Combo, Normal Skin' in x else 0)

# 1 if good for acne, else 0
skincare_df['is_good_for_acne'] = skincare_df['highlights_list'].apply(lambda x:1 if 'Good for: Acne/Blemishes' in x else 0)
# 1 if fragrance free, else 0
skincare_df['is_fragrance_free'] = skincare_df['highlights_list'].apply(lambda x:1 if 'Fragrance Free' in x else 0)

skincare_df[['product_name', 'is_for_dry_skin', 'is_for_oily_skin', 'is_good_for_acne', 'is_fragrance_free']].head(10)
```

```python
skincare_df.to_csv("data/processed/skincare_features.csv", index=False)
print("Saved!")
print(skincare_df.shape)
print(skincare_df.columns.tolist())
```
