```python
# Check how many unique highlights/skin type tag exist
import ast

# Get all highlights as lists and flatten
all_highlights = []
for val in skincare_df['highlights'].dropna():
    try:
        items = ast.literal_eval(val)
        all_highlights.extend(items)
    except:
        pass

# Most common highlights
import pandas as pd
pd.Series(all_highlights).value_counts().head(20)
```
